# Decomposition plots of figures 6,7 and supplementary figures 

## Import packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import colicycle.decomposition as dp
import tabulate
import numpy as np
```

## Define reference folder

```python
folder_prefix = '../'
tosave_folder = folder_prefix+'Plots/'
```

## Plots for Glycerol (auto) experimental data

### Load data and complete dataframe with initiation centric variables

```python
#load file
file_to_analyze = folder_prefix+'PreProcessed/20170327_GW339_temp/colidata.pkl'
datatype = 'exp'
time_scale = 3
colidata = pd.read_pickle(file_to_analyze)
```

```python
#define initiation centric variables
colidata = dp.cross_generation_construct(colidata, 'exp', time_scale = time_scale)
```

### Figure 7 Supp. 1

```python
plotname = 'fig_7_sup1'

coli_to_test, sorted_combinations, sorted_vals, table_to_plot = dp.decompose_double_adder(colidata)

fig = dp.plot_combinations_9array_v2(coli_to_test, sorted_combinations, sorted_vals, np.arange(0,9,1), dp.renaming_modelGW)
fig.suptitle('Double-adder Experimental', fontsize=35)
plt.show()

fig.savefig(tosave_folder+plotname+'.pdf')

table_to_plot = [dp.renaming_modelGW(x) for x in table_to_plot]
with open(tosave_folder+'tables/table_doubleadder_'+datatype+'.txt','w') as f:
    f.write(tabulate.tabulate(table_to_plot, tablefmt='latex_raw',headers=['combinations','independence']))

#keep only variables used in the plot, rename table and export
mapper ={}
tokeep_vars = np.unique(np.concatenate(sorted_combinations))
for k in tokeep_vars:
    mapper[k]=dp.renaming_modelGW([str(k)])[0].replace('$','').replace('\\','').replace('}','').replace('{','')

colidata_export = coli_to_test[tokeep_vars].copy()
colidata_export = colidata_export.rename(columns = mapper)
colidata_export.to_csv(folder_prefix+'Data_export/'+plotname+'.csv',index=False)
```

### Figure 6 Supp 1

```python
plotname = 'fig_6_sup1'
coli_to_test, sorted_combinations, sorted_vals, table_to_plot = dp.decompose_division_adder(colidata)

fig = dp.plot_combinations_9array3x3_v2(coli_to_test, sorted_combinations, sorted_vals, np.arange(0,9,1), dp.renaming_classic)
fig.suptitle('Division Adder Experimental', fontsize=35)
plt.show()

fig.savefig(tosave_folder+plotname+'.pdf')

#keep only variables used in the plot, rename table and export
mapper ={}
tokeep_vars = np.unique(np.concatenate(sorted_combinations))
for k in tokeep_vars:
    mapper[k]=dp.renaming_classic([str(k)])[0].replace('$','').replace('\\','').replace('}','').replace('{','')

colidata_export = coli_to_test[tokeep_vars].copy()
colidata_export = colidata_export.rename(columns = mapper)
colidata_export.to_csv(folder_prefix+'Data_export/'+plotname+'.csv',index=False)
```

### Figure 6A

```python
fig, ax  = dp.plot_combinations(coli_to_test, sorted_combinations, sorted_vals, 0, dp.renaming_classic, numvar = 3)
fig.savefig(tosave_folder+'fig_6_A1.pdf')
fig, ax  = dp.plot_combinations(coli_to_test, sorted_combinations, sorted_vals, 4, dp.renaming_classic, numvar = 3)
fig.savefig(tosave_folder+'fig_6_A2.pdf')
```
